```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("tips")
```

```python
sns.scatterplot(data=data,x='total_bill',y='tip')
sns.set_theme
plt.show()
```

```python
sns.scatterplot(data=data,x='total_bill',y='tip',hue="sex",style='time')
sns.set_theme

plt.show()
```

```python
sns.scatterplot(data=data,x='total_bill',y='tip',hue="day",style='sex')
plt.show()
# here how hue and style works but in this it is not good to read
```

```python
sns.scatterplot(data=data,x='total_bill',y='tip',hue="sex",style='day')
plt.show()
```
