```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("tips")
```

```python
sns.set_theme()
```

```python
sns.scatterplot(data=data,x='total_bill',y='tip')
plt.show()
```

```python
sns.scatterplot(data=data,x='total_bill',y='tip',hue="sex",style='time')
sns.set_theme

plt.show()
```

```python
sns.scatterplot(data=data,x='total_bill',y='tip',hue="day",style='sex')
plt.show()
# here how hue and style works but in this it is not good to read
```

```python
sns.scatterplot(data=data,x='total_bill',y='tip',hue="sex",style='day')
plt.show()
```

```python
sns.scatterplot(data=data,x='total_bill',y='tip',hue="sex",style='sex')
plt.show()
```

## Relplot

```python
tip = sns.load_dataset('tips')
tip.head()
```

```python
sns.relplot(data=tip,x='total_bill',y='tip',kind='line',col='sex')
```

```python
sns.relplot(data=tip,x='total_bill',y='tip',kind='line',col='sex',hue='smoker')
```

```python
sns.relplot(data=tip,x='total_bill',y='tip',kind='scatter',col='sex',hue='smoker')
```

```python
sns.relplot(data=tip,x='total_bill',y='tip',kind='scatter',col='sex',hue='smoker',row='time')
```

# Learning about resizing

```python
taxis = sns.load_dataset('taxis')
```

```python
plt.figure(figsize=(2,2)) # for controlling the size of plot
sns.relplot(
    data=taxis,
    x='total',
    y='tip',
    col='pickup_borough',
    hue='payment',
    row='color'
)
plt.show()
```

```python
sns.relplot(data=tip,x='total_bill',y='tip',kind='scatter',col='sex',hue='smoker',height=4,aspect=1)# aspect is just like width the width is cal by height*aspect
```
