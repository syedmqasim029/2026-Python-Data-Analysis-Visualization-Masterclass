### Categorica Scatter Plot

#### Strip plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.catplot(x="day", y="tip", data=tips,kind='strip',hue='day')
plt.show()
```

```python
sns.catplot(x="day", y="tip", data=tips,kind='strip',jitter=0.1,hue='day')
plt.show()
```

```python
sns.catplot(x="day", y="tip", data=tips,kind='swarm',hue='day')
plt.show()
```

```python
sns.catplot(x="day", y="tip", data=tips,kind='swarm',hue='sex')
plt.show()
```
