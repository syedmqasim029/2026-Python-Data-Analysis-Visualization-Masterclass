## Box plot 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
```

```python

plt.style.use("fivethirtyeight")
df.head()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(data=df["age"], orient="v", linewidth=3)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df["age"], orient="h", linewidth=3)
plt.show()
```

```python
sns.catplot(x="embark_town", y="age", data=df,kind='box',hue='alive',linewidth=3)
```
