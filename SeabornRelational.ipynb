## Scatter

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.style.use('fivethirtyeight')
```

```python
# sns.set_theme()
```

```python
da = sns.load_dataset('tips')
da.head()
```

```python

sns.relplot(x='total_bill', y='tip', data=da)
plt.show()
```

```python
# Hue
sns.relplot(x='total_bill',y='tip',data=da,hue='smoker')
```

```python
# Style
sns.relplot(x='total_bill',y='tip',data=da,style='smoker',hue='sex')
```

```python
# sizes
sns.relplot(data=da,x='total_bill',y='tip',hue='sex',style='smoker',size='size',sizes=(10,300))
```
